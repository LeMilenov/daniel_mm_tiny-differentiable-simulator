Display results in matplotlib

```python
import matplotlib.pyplot as plt

# Read data from file
with open("./grad_finite_data.txt", "r") as f:
    lines = f.readlines()

# Extract title, iteration, and cost values
titleFinite = lines[-1].strip()  # read the last line as the title
iterations_finite = []
costs_finite = []
for line in lines[:-1]:  # exclude the last line from the iteration and cost data
    iter, cost = line.strip().split(": ")
    iterations_finite.append(int(iter))
    costs_finite.append(float(cost))

# Read data from second file
with open("./grad_adam_data.txt", "r") as f:
    lines = f.readlines()

# Extract iteration and cost values
titleAdam = lines[-1].strip()  # read the last line as the title
iterations_adam = []
costs_adam = []
for line in lines[:-1]:
    iter, cost = line.strip().split(": ")
    iterations_adam.append(int(iter))
    costs_adam.append(float(cost))

# Plot data
plt.plot(iterations_finite, costs_finite,label="Finite Differences")
plt.plot(iterations_adam, costs_adam,label="Adam")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Cost vs Iteration")
plt.legend()
plt.show()
```
